# Clustering Test

```python
# synthetic classification dataset
from numpy import where
from sklearn.datasets import make_classification
from matplotlib import pyplot
import pandas as pd

# Import data
data = pd.read_excel('凸窗测试_全局.xlsx', sheet_name = 'Sheet1')
data = data.iloc[:50]
# data['高'] = 2880
nunique = data['标准件'].nunique()
print(nunique)

# Color data
data['颜色编号'] = data['标准件'].apply(lambda x: int(x.partition('件')[2]))

# Set X & Y values
X = data['宽']
Y = data['高']

# show the plot
pyplot.xlabel('宽')
pyplot.ylabel('高')
pyplot.scatter(X, Y, c = data['颜色编号'], cmap='tab20', s = 5)
pyplot.show()
```

```python
# affinity propagation clustering
from numpy import unique
from sklearn.cluster import AffinityPropagation
import pandas as pd

# define dataset
array = data[['宽', '高']].to_numpy()

# define the model
model = AffinityPropagation(damping=0.9)

# fit the model
model.fit(array)

# assign a cluster to each example
yhat = model.predict(array)

# retrieve unique clusters
clusters = unique(yhat)

# create scatter plot for samples from each cluster
for cluster in clusters:
	# get row indexes for samples with this cluster
	row_ix = where(yhat == cluster)
	# create scatter of these samples
	pyplot.scatter(array[row_ix, 0], array[row_ix, 1])

x = model.cluster_centers_[:,0] 
y = model.cluster_centers_[:,1] 
pyplot.scatter(x, y, c = 'black', marker = "x", s = 50)

# show the plot
pyplot.show()
```

```python
# agglomerative clustering
from sklearn.cluster import AgglomerativeClustering

# define dataset
array = data[['宽', '高']].to_numpy()

# define the model
model = AgglomerativeClustering(n_clusters=nunique)

# fit model and predict clusters
yhat = model.fit_predict(array)

# retrieve unique clusters
clusters = unique(yhat)

# create scatter plot for samples from each cluster
for cluster in clusters:
	# get row indexes for samples with this cluster
	row_ix = where(yhat == cluster)
	# create scatter of these samples
	pyplot.scatter(array[row_ix, 0], array[row_ix, 1])
    
# show the plot
pyplot.show()
```

```python
# birch clustering
from sklearn.cluster import Birch

# define dataset
array = data[['宽', '高']].to_numpy()

# define the model
model = Birch(threshold=0.01, n_clusters=nunique)

# fit the model
model.fit(array)

# assign a cluster to each example
yhat = model.predict(array)

# retrieve unique clusters
clusters = unique(yhat)

# create scatter plot for samples from each cluster
for cluster in clusters:
	# get row indexes for samples with this cluster
	row_ix = where(yhat == cluster)
	# create scatter of these samples
	pyplot.scatter(array[row_ix, 0], array[row_ix, 1])
# show the plot
pyplot.show()
```

```python
# dbscan clustering
from sklearn.cluster import DBSCAN

# define dataset
array = data[['宽', '高']].to_numpy()

# define the model
model = DBSCAN(eps=0.30, min_samples=5)

# fit model and predict clusters
yhat = model.fit_predict(array)

# retrieve unique clusters
clusters = unique(yhat)

# create scatter plot for samples from each cluster
for cluster in clusters:
	# get row indexes for samples with this cluster
	row_ix = where(yhat == cluster)
	# create scatter of these samples
	pyplot.scatter(array[row_ix, 0], array[row_ix, 1])
# show the plot
pyplot.show()
```
